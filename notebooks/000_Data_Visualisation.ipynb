```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_PATH = "../data/filtered_df.csv"
```

```python
df = pd.read_csv(DATA_PATH)
df
```

```python
df.disease.unique()
```

```python
df = df.sample(frac=1)
df
```

```python
df[["disease", "disease_encoded"]].head(4)
```

```python
tb_data = df[df["disease_encoded"] == 3].iloc[0]
tb_data
```

```python
non_tubercular_data = df[df["disease_encoded"] == 2].iloc[0]
non_tubercular_data
```

```python
non_infectious_data = df[df["disease_encoded"] == 1].iloc[0]
non_infectious_data
```

```python
dengue_data = df[df["disease_encoded"] == 0].iloc[0]
dengue_data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dengue_data.to_numpy()[0:1440])
plt.title("Dengue Data")
plt.xlabel("Time (in minutes)")
plt.ylabel("Temperature (normed)")
plt.savefig('figures/dengue_data.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(non_infectious_data.to_numpy()[0:1440])
plt.title("Non Infectious Disease Data")
plt.xlabel("Time (in minutes)")
plt.ylabel("Temperature (normed)")
plt.savefig('figures/non_infectious_data.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(non_tubercular_data.to_numpy()[0:1440])
plt.title("Non-tubercular Bacterial Infection Data")
plt.xlabel("Time (in minutes)")
plt.ylabel("Temperature (normed)")
plt.savefig('figures/non_tubercular_data.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tb_data.to_numpy()[0:1440])
plt.title("TB Data")
plt.xlabel("Time (in minutes)")
plt.ylabel("Temperature (normed)")
plt.savefig('figures/tb_data.pdf', bbox_inches='tight')
plt.show()
```
